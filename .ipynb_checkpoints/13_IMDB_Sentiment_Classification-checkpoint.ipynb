# IMDB Moview Review Sentiment Classification with Keras

Importing Packages

```python
from keras.preprocessing import sequence
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten, Lambda, SpatialDropout1D
from keras.layers.embeddings import Embedding
from keras.optimizers import SGD, RMSprop, Adam
from keras.layers.recurrent import LSTM
from keras.layers.convolutional import *
from keras.datasets import imdb
import matplotlib.pyplot as plt
import bcolz
import pickle
import numpy as np
import re
from numpy.random import random, permutation, randn, normal, uniform, choice
```

## Data
Review Data can be imported through Keras https://keras.io/datasets/#imdb-movie-reviews-sentiment-classification

Keras has also other datasets included: https://keras.io/datasets/

We only considering the top 5000 words

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(path="imdb.npz",
                                                      num_words=None,
                                                      skip_top=0,
                                                      maxlen=None,
                                                      seed=113,
                                                      start_char=1,
                                                      oov_char=2,
                                                      index_from=3)
```

```python
top_words = 5000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=top_words)

print('train sequences: {0}'.format(len(X_train)))
print('test sequences: {0}'.format(len(X_test)))
```

### Understanding the Encoding

Everything in this subsection is just for understanding the encoding of the data. No change in the data is happening here. Therefore just for the classification, you could skip this subsection.

The data is encoded as a sequence of word indexes. 

An index is coded based on the frequency, so index 3 is given to the word which is the third most frequent

Here's the 1st review. As you see, the words have been replaced by ids. The ids can be looked up in idx2word.

```python
X_train[0]
```

We will get the word indexes too so that we can decode the reviews. First we get from the dataset the dictionary **idx** with words as keys and indexes as values. 

```python
idx = imdb.get_word_index()
idx
```

When we sort the dictionary **idx** according the index, we see the words with a small index (aka the words that appear often)

```python
idx_arr = sorted(idx, key=idx.get)
idx_arr[:10]
```

```python
idx2word = {v: k for k, v in idx.items()}
idx2word
```

With the reverse dictionary **idx2word** we can now look up indexes and get the word. E.g. the index of the first word of the first review was 1415. What was the word?

```python
idx2word[1415]
```

Here's the whole review, mapped from ids to words. The **join** function adds between each element of a list a character (here a space) https://docs.python.org/3.6/library/stdtypes.html#str.join 

```python
' '.join([idx2word[index] for index in X_train[0] if index > 0])
```

### Sequence Cutting
we cut the review (sequence of word indexes) after 500 words. **pad_sequences** is a function of Keras https://keras.io/preprocessing/sequence/

Pad (with zero) or truncate each sentence to make consistent length.

```python
max_review_length = 500

X_train = sequence.pad_sequences(X_train, maxlen=max_review_length)
X_test = sequence.pad_sequences(X_test, maxlen=max_review_length)
```

This results in nice rectangular matrices that can be passed to ML algorithms. Reviews shorter than 500 words are pre-padded with zeros, those greater are truncated.

```python
X_test.shape
```

# Deep Learning Model 1: Dense

We create a Keras Sequential Model: https://keras.io/getting-started/sequential-model-guide/

You can add Layers (of Neurons) with **add**

* The first layer is an **Embedding** Layer of Keras 

    * https://keras.io/layers/embeddings/
    
* **Flatten**

    * This will create out of the matrix input a 1d-vector. Flatten does not change the data, it just rearranges the data. So a 2x4 Matrix will just be transformed into a 8 long vector.
    * https://keras.io/layers/core/#flatten
    
* The next layer is a fully connected **Dense** Layer. This means all nodes of the last layer are connected with all the nodes of the next layer.

    * https://keras.io/layers/core/
    
* Then we randomly drop (delete) 70% of the weights in the **Dropout** Layer. Dropout prevents overfitting.

    * https://keras.io/layers/core/#dropout
    
    * http://www.cs.toronto.edu/~rsalakhu/papers/srivastava14a.pdf 

* The last layer is a fully connected **Dense** Layer. This means all nodes of the last layer are connected with all the nodes of the next layer.

    * https://keras.io/layers/core/
    
* It has the **Sigmoid** activation function so that it is mapped between 0 and 1: 

    * https://keras.io/layers/core/#activation
    
    * https://en.wikipedia.org/wiki/Sigmoid_function

```python
# variable for the embedding. Will be used in the embedding step
embedding_vector_length = 32
print('Embedding Vector Length: {0}'.format(embedding_vector_length))
print('Number of Words (top_words): {0}'.format(top_words))
print('Maximal Length of Review: {0}'.format(max_review_length))
```

```python
model = Sequential()
model.add(Embedding(top_words, embedding_vector_length, input_length=max_review_length))
model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.7))
model.add(Dense(1, activation="sigmoid"))
```

After we specified the deep learning model, we have to compile it. 

* We use as the *loss function* **binary crossentropy**

    * https://keras.io/losses
    
* We use the optimization method **adam**

    * https://keras.io/optimizers/

* There are only two binary class: positive sentiment or negative sentiment

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Show the deep learning model.

```python
model.summary()
```

## Training

* We train the model with the method **fit**.

  * https://keras.io/models/sequential/#sequential-model-methods  

* **batch_size**: We let the model learn with 32 reviews each in one pass. One pass means

    1. Forward propagation
    2. Look at the error
    3. Backpropagation with adjusting of the weights

* **Epochs**: Number of time (iterations) that all the training data will go through the network (forward and backpropagation)

* **validation_data** You can either give a validation set or specify a **validation_split** in percent 

We save the accuracy and loss for each epoch in the variable **history**

```python
history = model.fit(X_train, y_train, batch_size=64, epochs=5, validation_data=(X_test, y_test))
```

This paper http://ai.stanford.edu/~amaas/papers/wvSent_acl2011.pdf cites that for the IMDB Dataset a state of the art accuracy of 0.883.

```python
# list all keys in history
print(history.history.keys())
```

```python
plt.plot(history.history['acc'], 'r', history.history['val_acc'], 'b')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

We see that the model in the test data gets nearly 100%. In the training data we get only 0.872 (after one epoch). Then the test accuracy will decrease. This shows overfitting.

```python
plt.plot(history.history['loss'], 'r', history.history['val_loss'], 'b')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

Similiar for the loss. The training data has nearly on loss (near 0). However, loss in the test data is increasing. Clear sign of overfitting.

## Evaluation


Because we used as the Evaluation set the total Test set, this next step is not necessary. However, we you used for evaluation another set or a split of the train set, than you can evaluate the model with **evaluate**

```python
scores = model.evaluate(X_test, y_test, batch_size=32)
print('\nEvaluation')
print('Test Loss: {0}'.format(scores[0]))
print('Test Accuracy: {0}'.format(scores[1]))
```

# Deep Learning Model 2: Long short-term memory (LSTM)


```python
# Same embedding parameter as before. Will be used in the embedding step
print('Embedding Vector Length: {0}'.format(embedding_vector_length))
print('Number of Words (top_words): {0}'.format(top_words))
print('Maximal Length of Review: {0}'.format(max_review_length))
```

We create a Keras Sequential Model: https://keras.io/getting-started/sequential-model-guide/

We save the new model to the same variable **model**. So we overwrite the last Dense model! If we would not like that, we would just use another variable name like lstm_model or what ever we like. 

You can add Layers (of Neurons) with **add**

* The first layer is an **Embedding** Layer of Keras 

    * https://keras.io/layers/embeddings/

* The second layer is a Long short-term memory (**LSTM**) Layer.

    * http://www.mitpressjournals.org/doi/pdf/10.1162/089976600300015015

* Then we randomly drop (delete) 50% of the weights in the **Dropout** Layer. Dropout prevents overfitting.

    * https://keras.io/layers/core/#dropout
    
    * http://www.cs.toronto.edu/~rsalakhu/papers/srivastava14a.pdf 

* The last layer is a fully connected **Dense** Layer. This means all nodes of the last layer are connected with all the nodes of the next layer.

    * https://keras.io/layers/core/

* It has the **Sigmoid** activation function so that it is mapped between 0 and 1: 

    * https://keras.io/layers/core/#activation
    
    * https://en.wikipedia.org/wiki/Sigmoid_function

```python
model = Sequential()
model.add(Embedding(top_words, embedding_vector_length, input_length=max_review_length))
model.add(LSTM(100))
model.add(Dense(1, activation="sigmoid"))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

## Training

```python
history = model.fit(X_train, y_train, batch_size=64, epochs=5, validation_data=(X_test, y_test))
```

The best result in the literature was 0.883. We get 0.874 at epoch 4. Not too bad :-)

## Create a plot of the Learning History

```python
plt.plot(history.history['acc'], 'r')
plt.plot(history.history['val_acc'], 'b')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'], 'r')
plt.plot(history.history['val_loss'], 'b')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

# Deep Learning Model 3: Convolutional Neural Network (CNN)

Instead of using **add** for adding one layer after another to the model, we can give the **Sequential** constructor a list with all the different layers. Most of the layers you alread know (Dense, Dropout, Embedding). We save the model in the variable conv_model (now we have to model variables: *model* and *conv_model*).

* **Convolution1d**: Because we have a one-dimensional sequence of words (our text) we need a 1d-Convolution. 

    * The first argument is the number of **filters** (here 64). Each filter has its own convolution and determines the dimensionality of the output space.
    * The second argument is the length of the convolution window (here 5). That means the convolution looks at sequences of 5 words each. 
    * **padding**: What is happening when our window (also called stride) touches the edge of our input. **same** means the input will just be filled with 0.
    * https://keras.io/layers/convolutional/
    * Image kernels (2d-convolution) explained visually: http://setosa.io/ev/image-kernels/
    * https://medium.com/@ageitgey/machine-learning-is-fun-part-3-deep-learning-and-convolutional-neural-networks-f40359318721
    * https://arxiv.org/abs/1311.2901
    
    
* **MaxPooling1d**
    
    * In the Pooling layer we will down-sample the data by reducing the dimensionality. In 1d-Pooling it seperates the input in non-overlapping sequences (windows), in 2d-Pooling in non-overlapping retangles. Then for each window or retangle only one number will be calculated. In MaxPooling this is the maximum of the values in the window or retangle.

    * https://keras.io/layers/pooling/
    * https://en.wikipedia.org/wiki/Convolutional_neural_network#Pooling_layer
    
* **Flatten**

    * This will create out of the input a 1d-vector
    * https://keras.io/layers/core/#flatten


```python
conv_model = Sequential([
    Embedding(top_words, embedding_vector_length, input_length=max_review_length),
    Dropout(0.2),
    Convolution1D(64, 5, padding='same', activation='relu'),
    Dropout(0.2),
    MaxPooling1D(),
    Flatten(),
    Dense(100, activation='relu'),
    Dropout(0.7),
    Dense(1, activation='sigmoid')])
```

```python
conv_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
conv_model.summary()
```

## Training

```python
history = conv_model.fit(X_train, y_train, batch_size=64, epochs=5, validation_data=(X_test, y_test))
```

## Ploting the history

```python
plt.plot(history.history['acc'], 'r')
plt.plot(history.history['val_acc'], 'b')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'], 'r')
plt.plot(history.history['val_loss'], 'b')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

The best result in the literature was 0.883. We see little overfitting and after the 2 epoch the accuracy in the test set goes down. After two epchs it was 0.8912. Higher than the best model! So we should stop the model after two epochs and use this model.
